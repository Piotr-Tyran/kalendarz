```python
from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///todo.db'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = False
db = SQLAlchemy(app)


class uzytkownicy(db.Model):

    __tablename__ = 'uzytkownicy'

    id_uzytkownika = db.Column(db.Integer, primary_key=True, nullable=False)
    email = db.Column(db.String(50), nullable=False, unique=True)
    username = db.Column(db.String(30), nullable=False)
    password = db.Column(db.String(100), nullable=False)
    id_uzyt_wyd = db.Column(db.Integer, db.ForeignKey('uzytkownik_wydarzenie.id'))
        
        
class wydarzenia(db.Model):

    __tablename__ = 'wydarzenia'

    id_wydarzenia = db.Column(db.Integer, primary_key=True, nullable=False)
    nazwa = db.Column(db.String(100), nullable=False)
    data_rozpoczecia = db.Column(db.Date, nullable=False)
    data_zakonczenia = db.Column(db.Date, nullable=False)
    typ = db.Column(db.Integer, nullable=False)
    id_uzyt_wyd = db.Column(db.Integer, db.ForeignKey('uzytkownik_wydarzenie.id'))
       
        
class uzytkownik_wydarzenie(db.Model):

    __tablename__ = 'uzytkownik_wydarzenie'

    id = db.Column(db.Integer, primary_key=True, nullable=False)
    czy_wlasciciel = db.Column(db.Boolean, nullable=False)
    czy_zakonczone = db.Column(db.Boolean, nullable=False)
    opis = db.Column(db.String(200), nullable=False)
    czy_zaakceptowane = db.Column(db.Boolean)
    id_uzyt = db.relationship('uzytkownicy', backref='uzytkownik_wydarzenie')
    id_wyd = db.relationship('wydarzenia', backref='uzytkownik_wydarzenie')
    id_tag = db.relationship('tagi', backref='uzytkownik_wydarzenie')
    id_przyp = db.relationship('przypomnienia', backref='uzytkownik_wydarzenie')
        
        
class tagi(db.Model):

    __tablename__ = 'tagi'

    id_tag = db.Column(db.Integer, primary_key=True, nullable=False)
    tagi = db.Column(db.String(50), nullable=False)
    id_uzyt_wyd = db.Column(db.Integer, db.ForeignKey('uzytkownik_wydarzenie.id'))
    
        
class przypomnienia(db.Model):

    __tablename__ = 'przypomnienia'

    id_przyp = db.Column(db.Integer, primary_key=True)
    czas_przed = db.Column(db.Date, nullable=False)
    id_uzyt_wyd = db.Column(db.Integer, db.ForeignKey('uzytkownik_wydarzenie.id'))


             

db.create_all()
```
